```python


import pandas as pd
import plotly.graph_objects as go

# File paths - update these if running locally
files = {
    'Brazil': '3_Jupyter_Sankey_BRA.xlsx',
    'DR Congo': '3_Jupyter_Sankey_COD.xlsx',
    'Indonesia': '3_Jupyter_Sankey_IDN.xlsx'
}

def load_country_data(filepath):
    """Load and prepare data for Sankey diagram"""
    df = pd.read_excel(filepath)
    all_nodes = df['Node, Label'].dropna().unique()
    
    sources = []
    targets = []
    values = []
    link_colors = []
    
    for _, row in df.iterrows():
        if pd.notna(row['Source']) and pd.notna(row['Target']) and pd.notna(row['Value']):
            sources.append(int(row['Source']))
            targets.append(int(row['Target']))
            values.append(row['Value'])
            link_colors.append(row['Link Color'] if pd.notna(row['Link Color']) else 'rgba(200, 200, 200, 0.4)')
    
    node_colors = []
    for node in all_nodes:
        node_row = df[df['Node, Label'] == node].iloc[0] if len(df[df['Node, Label'] == node]) > 0 else None
        if node_row is not None and pd.notna(node_row['Color']):
            node_colors.append(node_row['Color'])
        else:
            node_colors.append('#cccccc')
    
    return {
        'nodes': all_nodes.tolist(),
        'node_colors': node_colors,
        'sources': sources,
        'targets': targets,
        'values': values,
        'link_colors': link_colors
    }

# Load data for all countries
data = {}
for country, filepath in files.items():
    data[country] = load_country_data(filepath)

# Create figure
fig = go.Figure()

# Define vertical positions for each diagram
y_positions = [
    [0.66, 0.94],  # Brazil (top)
    [0.36, 0.64],  # DR Congo (middle)
    [0.06, 0.34]   # Indonesia (bottom)
]

# Add each Sankey diagram
for idx, (country, country_data) in enumerate(data.items()):
    fig.add_trace(go.Sankey(
        domain=dict(
            x=[0, 1],
            y=y_positions[idx]
        ),
        orientation="h",
        valueformat=".0f",
        node=dict(
            pad=50,
            thickness=25,
            line=dict(color="black", width=0),
            label=country_data['nodes'],
            color=country_data['node_colors']
        ),
        link=dict(
            source=country_data['sources'],
            target=country_data['targets'],
            value=country_data['values'],
            color=country_data['link_colors']
        ),
        textfont=dict(size=15, family="Arial")
    ))

# Add country labels
annotations = [
    dict(
        text="<b>Brazil</b>",
        x=0.5, y=0.97,
        xref="paper", yref="paper",
        showarrow=False,
        font=dict(size=18, family="Arial", color="black"),
        xanchor="center"
    ),
    dict(
        text="<b>DR Congo</b>",
        x=0.5, y=0.67,
        xref="paper", yref="paper",
        showarrow=False,
        font=dict(size=18, family="Arial", color="black"),
        xanchor="center"
    ),
    dict(
        text="<b>Indonesia</b>",
        x=0.5, y=0.37,
        xref="paper", yref="paper",
        showarrow=False,
        font=dict(size=18, family="Arial", color="black"),
        xanchor="center"
    )
]

# Update layout with unified settings
fig.update_layout(
    title=dict(
        text="<b>Transfer of Points for SDGs 8.4.1 and 12.2.1<br>by Area of Origin, Transfer and Destination</b>",
        x=0.5,
        xanchor='center',
        font=dict(size=22, family="Arial")
    ),
    height=2480,
    width=1754,
    font=dict(size=15, family="Arial"),
    showlegend=False,
    paper_bgcolor='white',
    annotations=annotations,
    margin=dict(t=120, b=50, l=80, r=80)
)

# Save outputs
print("Saving HTML file...")
fig.write_html('combined_sankey_diagrams.html')
print("✓ HTML saved: combined_sankey_diagrams.html")

# Try to save as PNG (requires kaleido and Chrome)
try:
    import kaleido
    print("Attempting PNG export...")
    fig.write_image('combined_sankey_diagrams.png', width=1754, height=2480, scale=2)
    print("✓ PNG saved: combined_sankey_diagrams.png")
except ImportError:
    print("! PNG export requires kaleido: pip install kaleido")
except Exception as e:
    print(f"! PNG export failed: {e}")
    print("  You may need to install Chrome or use the HTML file instead")

# Show the figure
fig.show()

print("\n✓ Done! The combined diagram has:")
print("  • Unified font size (15pt) and family (Arial)")
print("  • Consistent node padding and thickness")
print("  • Fixed dimensions: 1754 × 2480 pixels")
print("  • All three diagrams stacked vertically")
```
